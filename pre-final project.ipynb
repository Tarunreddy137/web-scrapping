# Scraping Data from Zigwheels

#### Importing required libraries

```python
import requests 
from bs4 import BeautifulSoup
```

```python
# requests library is used scarp the data from URL
url='https://www.zigwheels.com/newcars/Skoda'
page=requests.get(url)
page
page.url
page.text
soup = BeautifulSoup(page.text)
#soup library is used to call the request library
a=soup.find_all('div',class_="p-15 pt-10 mke-ryt rel")
a

cars = []
for i in a:
    cars.append(i.text)
cars

#scarping the data for brands
Brands = []
for i in soup.find_all("div",class_="p-15 pt-10 mke-ryt rel"):
    a = i.text
    Brands.append(a.split("\t")[0].strip().split()[0])
Brands

#scarping the data for carname
carname = []
for i in soup.find_all("div",class_="p-15 pt-10 mke-ryt rel"):
    a = i.text
    carname.append(a.split("\t")[0].strip().split()[1])
carname
#scarping the data for prices
prices= []
for i in soup.find_all("div",class_="p-15 pt-10 mke-ryt rel"):
    a = i.text
    prices.append(a.split("\t")[1].strip())
prices
#scarping the data for Engine_cc
Engine_cc = []
for i in soup.find_all("div",class_="p-15 pt-10 mke-ryt rel"):
    a = i.text
    Engine_cc.append(a.split("\t")[2].strip().replace("\n",' ').replace('|',' ').split()[3])
Engine_cc
#scraping the data for Mileage
Mileage=[]
for i in soup.find_all("div",class_="p-15 pt-10 mke-ryt rel"):
    a = i.text
    Mileage.append(a.split("\t")[2].strip().replace("\n",' ').replace('|',' ').split()[5])
Mileage
#scarping the data for version
version=[]
for i in soup.find_all("div",class_="p-15 pt-10 mke-ryt rel"):
    a = i.text
    version.append(a.split("\t")[2].strip().replace("\n",' ').replace('|',' ').split()[7])
version
b=soup.find_all('a',class_='r-wn w')
b
vechiles = []
for i in b:
    vechiles.append(i.text)
vechiles
vechiles[0].replace('\n','').replace('|','').split()[0]
vechiles[0].replace('\n','').replace('|','').split()[1]
#scarping data for Ratings
Ratings=[]
for i in b:
    a = i.text
    Ratings.append(a.replace('\n','').replace('|','').split()[0])
Ratings
#scarping data for Reviews
Reviews=[]
for i in b:
    a = i.text
    Reviews.append(a.replace('\n','').replace('|','').split()[1])
Reviews
# creating a dictionary
import pandas as pd
dictionary = {'Brands':Brands,'carname':carname,'prices':prices,'Engine_cc':Engine_cc,'Mileage':Mileage,'version':version,'Ratings':Ratings,'Reviews':Reviews}
dictionary
Skoda = pd.DataFrame(dictionary)
Skoda
```

```python
Skoda.to_csv('Slavia.csv')
```

# Data cleaning

```python
final_df['Engine_cc'].unique()
```

```python
final_df['Engine_cc']=final_df['Engine_cc'].replace('On-road','').replace('June','').replace('Electric','').replace('Petrol','').replace('41.5bhp@4000rpm','').replace('20.8-19.7kWh/100km','').replace('500km','').replace('493-590Km','').replace('372-425km',).replace('312km','').replace('388452Km','')
```

```python
final_df['Mileage'].unique()
```

```python
final_df['Mileage']=final_df['Mileage'].replace('Petrol','').replace('Diesel','').replace('Electric','').replace('View','').replace('June','').replace('CNG','').replace('cc','').replace('455-471km','').replace('Price','').replace('Offers','').replace('kmpl','')
```

# 

```python
final_df['Reviews'].unique()
```

```python
final_df['Reviews']=final_df['Reviews'].replace('read','').replace('nan','')
```

```python
final_df['version'].unique()
```

```python
final_df['version']=final_df['version'].replace('June','Petrol').replace('View','Diesel').replace('Offers','Diesel').replace('nan','Diesel').replace('CNG','Diesel')
```

```python
final_df['Ratings'].unique()
```

```python
final_df['Ratings']=final_df['Ratings'].replace('nan','')
```

```python
final_df['prices'].unique()
```

```python
final_df['prices']=final_df['prices'].replace('Price To Be Announced','')
```
